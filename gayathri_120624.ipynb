```python
# Install necessary libraries
!pip install python-docx

# Import necessary libraries
import docx

# Load the document
file_path = '/content/sample_data/FoodCourt.docx'
doc = docx.Document(file_path)

# Extract text from each paragraph
reviews = [para.text for para in doc.paragraphs if para.text]

# Display the extracted reviews
for review in reviews:
    print(review)
```

```python
import re
import string

# Preprocess text function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters
    text = re.sub(r'[^\w\s]', '', text)
    return text

# Apply preprocessing to each review
cleaned_reviews = [preprocess_text(review) for review in reviews]

# Display cleaned reviews
for review in cleaned_reviews:
    print(review)
```

```python
!pip install spacy
!python -m spacy download en_core_web_sm

import spacy

# Load SpaCy model
nlp = spacy.load('en_core_web_sm')

# Function to extract aspects
def extract_aspects(text):
    doc = nlp(text)
    aspects = []
    for chunk in doc.noun_chunks:
        aspects.append(chunk.text)
    return aspects

# Apply aspect extraction to each review
extracted_aspects = [extract_aspects(review) for review in cleaned_reviews]

# Display extracted aspects
for review, aspects in zip(reviews, extracted_aspects):
    print(f"Review: {review}\nAspects: {aspects}\n")
```

```python
# Define keywords for aspects
aspects_keywords = {
    'ordering process': ['ordering', 'order process'],
    'variety of restaurants': ['variety', 'restaurants', 'choices'],
    'interface': ['interface', 'navigation', 'menus'],
    'delivery times': ['delivery', 'times'],
    'glitches': ['glitches', 'bugs', 'crashes'],
    'customizing orders': ['customizing', 'options', 'dietary'],
    'accuracy': ['accuracy', 'correct items'],
    'notifications': ['notifications', 'updates'],
    'rewards': ['rewards', 'discounts', 'loyalty'],
    'customer service': ['customer service', 'support'],
    'past orders': ['past orders', 'reorder'],
    'payment': ['payment', 'secure'],
    'live tracking': ['live tracking', 'tracking'],
    'promotions': ['promotions', 'discounts'],
    'user reviews': ['user reviews', 'ratings'],
    'contact delivery driver': ['contact delivery driver', 'driver'],
    'updated menus': ['updated menus', 'menus'],
}

# Function to identify aspects
def identify_aspects(text):
    identified_aspects = {}
    for aspect, keywords in aspects_keywords.items():
        if any(keyword in text for keyword in keywords):
            identified_aspects[aspect] = text
    return identified_aspects

# Apply aspect extraction
aspects = [identify_aspects(review) for review in cleaned_reviews]

# Display identified aspects
for review, aspect in zip(reviews, aspects):
    print(f"Review: {review}\nAspects: {aspect}\n")
```

```python
import numpy as np

# Function to categorize sentiments
def categorize_sentiment(sentiment):
    if sentiment['compound'] >= 0.05:
        return 'positive'
    elif sentiment['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Determine overall sentiment
compound_scores = [sentiment['compound'] for sentiment in sentiments]
overall_sentiment_score = np.mean(compound_scores)
overall_sentiment = categorize_sentiment({'compound': overall_sentiment_score})

# Compile positives, negatives, and scopes for improvement
positive_aspects = []
negative_aspects = []

for review, aspect, sentiment in zip(reviews, aspects, sentiments):
    sentiment_category = categorize_sentiment(sentiment)
    if sentiment_category == 'positive':
        positive_aspects.append(aspect)
    elif sentiment_category == 'negative':
        negative_aspects.append(aspect)

# Identify scopes for improvement from negative aspects
scopes_for_improvement = []
for aspect in negative_aspects:
    for key in aspect:
        scopes_for_improvement.append(key)

# Structure the summary
summary = {
    "App Name": "FoodCourt",
    "Overall Sentiment": overall_sentiment,
    "Positives": positive_aspects,
    "Negatives": negative_aspects,
    "Scopes for Improvement": list(set(scopes_for_improvement))
}

# Display summary
import pprint
pprint.pprint(summary)
```

```python
# Open the text file
file_path = '/content/sample_data/WhiteHatJr.txt'  # Provided file path
with open(file_path, 'r', encoding='utf-8') as file:
    data = file.readlines()

# Initialize lists to store reviews
reviews = []

# Parse the data
current_review = ""
for line in data:
    line = line.strip()
    if line:
        if line.isupper():  # Assuming company names are in uppercase
            # Start new review
            current_review = ""
        else:
            # Append review text
            current_review += line + " "
            # Store review if it's complete (assuming reviews end with punctuation)
            if line.endswith('.') or line.endswith('!') or line.endswith('?'):
                reviews.append(current_review.strip())

# Print the number of reviews and sample data
print("Total reviews:", len(reviews))
print("\nSample review:", reviews[121])
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon if not already downloaded
nltk.download('vader_lexicon')

# Initialize the VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Analyze sentiment for each review
sentiments = []
for review in reviews:
    sentiment_score = sid.polarity_scores(review)
    sentiment = 'positive' if sentiment_score['compound'] > 0 else 'negative' if sentiment_score['compound'] < 0 else 'neutral'
    sentiments.append(sentiment)

# Print the sentiments for the first few reviews
for i in range(5):
    print("Review:", reviews[i])
    print("Sentiment:", sentiments[i])
    print()
```

```python
from collections import Counter

# Calculate overall sentiment
sentiment_counts = Counter(sentiments)
overall_sentiment = max(sentiment_counts, key=sentiment_counts.get)

# Identify positive and negative aspects
positive_reviews = [review for review, sentiment in zip(reviews, sentiments) if sentiment == 'positive']
negative_reviews = [review for review, sentiment in zip(reviews, sentiments) if sentiment == 'negative']

# Print overall sentiment
print("Overall Sentiment:", overall_sentiment)
print()

# Print positive aspects
print("Positive Aspects:")
for i, review in enumerate(positive_reviews[:3]):  # Print the first few positive reviews
    print(f"Positive Review {i+1}:")
    print(review)
    print()

# Print negative aspects
print("Negative Aspects:")
for i, review in enumerate(negative_reviews[:3]):  # Print the first few negative reviews
    print(f"Negative Review {i+1}:")
    print(review)
    print()
```
